### Uniform geographical grid clustering (GC)

```python
cell = (6, 5)
lat, lon = station['latitude'], station['longitude']
y1, y2, x1, x2 = lat.min(), lat.max(), lon.min(), lon.max()
dy, dx = (y2-y1+0.001)/cell[0], (x2-x1+0.001)/cell[1]
iy = list(map(lambda i:(lat>=y1+i*dy)&(lat<y1+(i+1)*dy), range(cell[0])))
ix = list(map(lambda i:(lon>=x1+i*dx)&(lon<x1+(i+1)*dx), range(cell[1])))

_gclusters, cnt, s = [], 0, 0
station['gcluster'] = -1
for y in iy:
    for x in ix:
        c = station[y & x]
        s += len(c)
        if len(c) > 0:
            station.loc[y & x, 'gcluster'] = cnt
            _gclusters += [c]
            cnt += 1
print(len(station), s)
print(station.head())
plt.figure(figsize=(7,7))
for c in _gclusters:
    plt.scatter(c['longitude'], c['latitude'])
```

#### Transit-matrix generation

**Maximum likelyhood estimation**

the pmf of **multinormial distribution**: 
$$f(\boldsymbol{X},N;\boldsymbol{\theta})=
\binom{N}{x_1,x_2,\cdots,x_n}
\theta_1^{x_1}\theta_2^{x_2}\cdots\theta_n^{x_n}
$$
here, $
\boldsymbol{X}=(x_1,x_2,\cdots,x_n),
\boldsymbol{\theta}=(\theta_1,\theta_2,\cdots,\theta_n),
\sum\theta_j=1
$

for each sample $\boldsymbol{X}^i$ checks in at Cluster $C_k$, 
$$N=1\quad x_k^i=1\qquad x_j^i=0, \forall j \ne k$$

suppose the number of the samples that check in at Clusters
$(C_1,C_2,\cdots,C_m)$ is $(l_1, l_2, \cdots, l_m)$ respectively.

the likelyhood function:
$$L(\boldsymbol{\theta})=
\prod_i^{T}p(\boldsymbol{x}^i,1;\boldsymbol{\theta})=\prod_{j=1}^m\theta_j^{l_j}$$
$$\ln L(\boldsymbol{\theta})
%=\sum_i^{T}\ln p(\boldsymbol{X}_i;\boldsymbol{\theta})
=\sum_j^ml_j\ln\theta_j
$$
Now maximize 
$\quad\ln L(\boldsymbol{\theta})
\quad\mathrm{s.t.}\ \sum\theta_j=1
$

by applying **Lagrange multiplier method**

let $F(\boldsymbol{\theta},\lambda)=\ln L(\boldsymbol{\theta})-\lambda g(\boldsymbol{\theta}),\quad
g(\boldsymbol{\theta})=\sum\theta_j-1$

$$\begin{aligned}
&\frac{\partial F(\boldsymbol{\theta},\lambda)}{\partial\theta_j}=\frac{l_j}{\theta_j}-\lambda=0\\
&\theta_j=\lambda^{-1}l_j\\
\therefore\quad&\theta_j\propto l_j\\
&\theta_k=\frac{l_k}{\sum l_j}
\end{aligned}$$

```python
def calc_slot(raw_date_str):
    dt = datetime.strptime(raw_date_str, '%m/%d/%Y %H:%M')
    weekday = dt.weekday()
    hr = dt.hour
    if weekday < 5:
        if hr >= 7:
            if hr < 11:
                return 0
            if hr < 16:
                return 1
            if hr < 21:
                return 2
        return 3
    if hr < 9:
         return 4
    if hr < 19:
        return 5
    return 6
def get_t_vector(station_id):
    st = bc_df[bc_df['start station id'] == station_id][:]
    st['slot'] =  st['starttime'].map(calc_slot)
    #st = st.drop('starttime', axis=1)
    by_slots = dict(st.groupby('slot').groups)
    for key in by_slots.keys():
        ndict = dict(st.loc[by_slots[key]].groupby('bcluster').size())
        count = np.array(list(map(lambda x:ndict[x] if x in ndict else 0, range(K1))))
        count = count / count.sum()
        by_slots[key] = count
    if len(by_slots) < 7:
        mean = np.array(list(by_slots.values())).mean(axis=0)
        for i in range(7):
            if i not in by_slots.keys():
                by_slots[i] = mean
    return np.array(list(by_slots.values())).reshape(7*K1)
```

```python
def sameCluster(s, s0):
    A = sorted(s.groupby('bcluster').groups.values(), key = lambda x: x[0])
    B = sorted(s0.groupby('bcluster').groups.values(), key = lambda x: x[0])
    if len(A) != len(B):
        return False
    for a, b in zip(A, B):
        if len(a) != len(b):
            return False
        if (a != b).sum() != 0:
            return False
    return True
```

```python
n = len(station)
K1, K2, K = 24, 8, 5
X = np.array(station[['latitude', 'longitude']])
station['bcluster'] = KMeans(n_clusters=K1).fit(X).labels_
for k in range(K):
    bc_df = nyc_df[['starttime','start station id','end station id']]
    bc_df = bc_df.merge(station[['id', 'bcluster']],
                how='left',left_on='end station id',right_on='id').drop('id',axis=1)
    last_station = station[['id', 'bcluster']]
    X = np.array(list(station['id'].map(get_t_vector)))
    station['bcluster'] = KMeans(n_clusters=K2).fit(X).labels_
    vals = station.groupby('bcluster').groups.values()
    offset = 0
    for v in vals:
        X = np.array(station.loc[v][['latitude', 'longitude']])
        n_c = int(np.round(len(v)*K1/n))
        station.loc[v,'bcluster'] = KMeans(n_clusters=n_c).fit(X).labels_ + offset
        offset += n_c
    if sameCluster(station, last_station):
        break
K1 = offset # offset might be K1 + 1 because of rounding error
plt.figure(figsize=(7,7))
_cluster = list(station.groupby('bcluster').groups.values())
for c in _cluster:
    plt.scatter(station.loc[c]['longitude'], station.loc[c]['latitude'])
plt.title('BC')
```

```python
tmp_df = nyc_df[['starttime','start station id']]
tmp_df = tmp_df.merge(station[['id', 'bcluster']],
            how='left',left_on='start station id',right_on='id').drop('id',axis=1)
tmp_df['sdt'] = tmp_df['starttime'].map(lambda x:datetime.strptime(x, '%m/%d/%Y %H:%M'))
tmp_df['year'] = tmp_df['sdt'].map(lambda x:x.year)
tmp_df['month'] = tmp_df['sdt'].map(lambda x:x.month)
tmp_df['day'] = tmp_df['sdt'].map(lambda x:x.day)
tmp_df['hour'] = tmp_df['sdt'].map(lambda x:x.hour)

by_ymdh = dict(tmp_df.groupby(['year','month','day','hour']).groups)
cnt_dict = {}
for key in by_ymdh.keys():
    ndict = dict(tmp_df.loc[by_ymdh[key]].groupby('bcluster').size())
    count = np.array(list(map(lambda x:ndict[x] if x in ndict else 0, range(K1))))
    cnt_dict[key] = count
```

```python
params = {
    'rou': [0.7, 0.8],
    'alpha': [0.6, 0.2, 0.05, 0.3, 0.1, 0.8],
    'sigma': [start_out_df['Temperature ° F'].std(), start_out_df['Wind Speed mph'].std()],
    'H': 48,
    'J': 4,
    'q': 1.2
}
```

```python
def weather_sim(w1, w2):
    ws = [
        [1,  params['alpha'][0], params['alpha'][1], params['alpha'][2]],
        [params['alpha'][0], 1, params['alpha'][3], params['alpha'][4]],
        [params['alpha'][1], params['alpha'][3], 1, params['alpha'][5]],
        [params['alpha'][2], params['alpha'][4], params['alpha'][5], 1],
    ]
    return ws[w1][w2]
#     if w1 == w2:
#         return 1
#     if w1 > w2:
#         return lam_2_func(w2, w1)
#     # w1 < w2:
#     s = [0, 3, 5]
#     return params['alpha'][s[w1]+ w2 - w1 - 1]

def sim_w(f1, f2):
    if f1['isweekend'] != f2['isweekend']:
        return 0
    
    delta_h = np.abs(f1['hour'] - f2['hour'])
    if delta_h > 12:
        delta_h = 24 - delta_h
    delta_d = np.abs(f1['day'] - f2['day'])
    lam_1 = (params['rou'][0] ** delta_h) * (params['rou'][1] ** delta_d)
    
    lam_2 = weather_sim(f1['weather'], f2['weather'])
    
    temp = (f1['Temperature ° F'] - f2['Temperature ° F']) / params['sigma'][0]
    temp *= temp
    wv = (f1['Wind Speed mph'] - f2['Wind Speed mph']) / params['sigma'][1]
    wv *= wv
    Ker = np.exp(-temp-wv)/(2*np.pi*params['sigma'][0]*params['sigma'][1])
    
    return lam_1*lam_2*Ker

def predict_proportion(data, t):
    w = [sim_w(data.loc[i], data.loc[t]) for i in range(t-params['H'], t)]
    w = np.array([w])
    w = w / w.sum()
    Prop = [data.loc[i, 'proportion'] for i in range(t-params['H'], t)]
    Prop = np.array(Prop)
    return np.dot(w, Prop)

def predict_proportion_ex(data, t):
    e = [data.loc[i,'proportion'] - predict_proportion(data,i) for i in range(t-params['J'],t)]
    e = np.array(e)
    e = e.reshape(e.shape[0],e.shape[-1])
    q = [params['q'] ** i for i in range(params['J'])]
    q = np.array([q])
    q = q / q.sum()
    return predict_proportion(data, t)  + np.dot(q, e)
```

## Inter-cluster Transition Learning

$T_{t,C_i,C_j}$ is a transition probability from cluster $C_i$ to cluster $C_j$ in time $t$.

$$T_{t,C_i,C_j} = Pr(\text{check in at } C_j | \text{ check out at } C_i \text{ in time } t)$$


Each row is a margin distribution.

So the sum of each row in matrix T is 1. 

```python
tmp_df = nyc_df[['starttime','start station id', 'end station id']]
tmp_df = tmp_df.merge(station[['id', 'bcluster']],
            how='left',left_on='start station id',right_on='id').drop('id',axis=1)
tmp_df = tmp_df.merge(station[['id', 'bcluster']],
            how='left',left_on='end station id',right_on='id').drop('id',axis=1)
tmp_df.head()
```

```python
tmp_df['sdt'] = tmp_df['starttime'].map(lambda x:datetime.strptime(x, '%m/%d/%Y %H:%M'))
tmp_df['year'] = tmp_df['sdt'].map(lambda x:x.year)
tmp_df['month'] = tmp_df['sdt'].map(lambda x:x.month)
tmp_df['day'] = tmp_df['sdt'].map(lambda x:x.day)
tmp_df['hour'] = tmp_df['sdt'].map(lambda x:x.hour)

by_ymdh = dict(tmp_df.groupby(['year','month','day','hour']).groups)
tranmat = {}
for key in by_ymdh.keys():
    start = dict(tmp_df.loc[by_ymdh[key]].groupby('bcluster_x').groups)
    for s_key in start.keys():
        end = dict(tmp_df.loc[start[s_key]].groupby('bcluster_y').size())
        count = np.array(list(map(lambda x:end[x] if x in end else 0, range(K1))))
        start[s_key] = count / count.sum()
    if len(start) < K1:
        for i in range(K1):
            if i not in start.keys():
                start[i] = np.zeros(K1)
    tranmat[key] = np.array(list(start.values()))
tranmat_df = pd.DataFrame({'translation matrix': pd.Series(tranmat)})
tranmat_df.head()
```

### P-TD

$$
p(x)
=\frac{1}{\sqrt{2\pi}\sigma x}
\mathrm{e}^{-\frac{(\ln x-\mu)^2}{2\sigma^2}}
\quad (x > 0)
$$

$$
\begin{aligned}
\int_a^bp(x)\mathrm{d}x
=&\frac{1}{\sqrt{2\pi}\sigma}\int_a^b
\mathrm{e}^{-\frac{(\ln x-\mu)^2}{2\sigma^2}}\mathrm{d}\ln x\\
=&\frac{1}{\sqrt{2\pi}\sigma}\int_\alpha^\beta
\mathrm{e}^{-\frac{(y-\mu)^2}{2\sigma^2}}\mathrm{d}y\\
=&\Phi_{\mu,\sigma}(\beta)-\Phi_{\mu,\sigma}(\alpha)\\
=&\Phi_{\mu,\sigma}(\ln b)-\Phi_{\mu,\sigma}(\ln a)\\
=&\Phi_{0,1}(\frac{\ln b-\mu}{\sigma})-\Phi_{0,1}(\frac{\ln a-\mu}{\sigma})
\end{aligned}
$$

```python
online_df = nyc_df[['starttime','stoptime','start station id','end station id']]
online_df = online_df.merge(station[['id', 'bcluster']],
            how='left',left_on='start station id',right_on='id').drop('id',axis=1)
online_df['startdt'] = online_df['starttime'].map(lambda x:datetime.strptime(x, '%m/%d/%Y %H:%M'))
online_df['stopdt'] = online_df['stoptime'].map(lambda x:datetime.strptime(x, '%m/%d/%Y %H:%M'))
online_df['starthour'] = online_df['startdt'].map(lambda x:(x.day-1)*24+x.hour)
online_df['stophour'] = online_df['startdt'].map(lambda x:(x.day-1)*24+x.hour)
online_df = online_df[['starthour','stophour','bcluster']]
online_df.head(1000).tail()
```

```python
def D_int(a, b, i, j):
    norm = scipy.stats.norm(Mu[i, j],Sigma[i, j])
    if a:
        return norm.cdf(np.log(b*60)) - norm.cdf(np.log(a*60))
    if b:
        return norm.cdf(np.log(b*60))
    return 0.0

def p_td(t):
    active = online_df[(online_df.starthour<=t) & (online_df.stophour>=t)]
    longtrip = active[active.starthour < t]
    shorttrip = active[active.starthour == t]
    O = shorttrip.groupby('bcluster').size()
    E = np.zeros(K1)
    for i in range(K1):
        T = [
            tranmat_df.loc[t,'translation matrix'][j][i]
            for j in longtrip['bcluster']
        ]
        D = [
            D_int((t-x)*60, (t-x+1)*60, i, j)
            for x,j in zip(longtrip['starthour'], longtrip['bcluster'])
        ]
        e1 = (np.array(T) * np.array(D)).sum()
        e2 = 0
        for dt in range(1,61):
            for j in range(K1):
                try:
                    e2 += O[j]/60*tranmat_df.loc[t,'translation matrix'][j][i]*D_int(0,60-dt,i,j)
                except:
                    pass
        E[i] = e1 + e2
    return E
```
